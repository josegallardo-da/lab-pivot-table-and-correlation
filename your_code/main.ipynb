#### Open the ` Employee.csv` file and name your dataset `employee`

```python
# your answer here

employee = pd.read_csv('Employee.csv')
employee
```

#### What's the mean salary by department. 
##### Tip : There are 2 ways to do it - one way is with and the other way is without pivot tables. Do it both ways. 


```python
# your answer here-1st way

mean = employee.groupby('Department').agg({'Salary':'mean'})
mean
```

```python
# your answer here-2nd way

mean2 = pd.pivot_table(employee,values='Salary',index='Department', aggfunc={'Salary':np.mean})
mean2
```

#### We want now to group the data by more than one column. Use Pandas pivot table function to select the mean salary by department and title

```python
# your answer here

salaries = pd.pivot_table(employee,values='Salary',index='Department',columns = 'Title', aggfunc={'Salary':np.mean})
salaries
```

#### We want to display the number of years of work experience and the salary for each Title. Add Years as a column in your pivot table
*Hint: Use Pandas doc(see References in README.md) to explore the pivot_table() arguments*

```python
# your answer here

years = pd.pivot_table(employee,values='Salary',index='Title', columns = 'Years', aggfunc={'Salary':np.mean}).fillna('')
years
```

#### We want to know the information about the mean salary and number of employees in each department and for each job title

```python
# your answer here

titles = pd.pivot_table(employee,values=['Salary','Name'],index='Department', columns = 'Title', aggfunc={'Salary':np.mean,'Name':'count'}).fillna('')
titles
```

#### We want to know for each department and for each job title the Salary's median,minimum,maximum and standard deviation. Comment your results

```python
# your answer here

sal_department = pd.pivot_table(employee,values=['Salary'],index=['Department','Title'], aggfunc={'Salary':['median','max','min',np.std]})
sal_department
```

#### Based on your comments, fill the missing values with an appropriate value

```python
# your answer here

pd.pivot_table(employee,values=['Salary'],index=['Department','Title'], aggfunc={'Salary':['median','max','min',np.std]},fill_value=0)
```

#### The stake holders want to know for each department the number of employees and how much money is spend on salaries. Could you provide that information? 

```python
# your answer here

stake_holders = pd.pivot_table(employee,values=['Salary'],index='Department', columns = 'Title', aggfunc={'Salary':['sum','count']}).fillna(0)
stake_holders
```

#### For each Department and Title, we want to know the the total years of work experience and the mean salary

```python
# your answer here

dep_title = pd.pivot_table(employee,values=['Salary','Years'],index='Department', columns = 'Title', aggfunc={'Salary':'mean','Years':'mean'}).fillna(0)
dep_title
```
